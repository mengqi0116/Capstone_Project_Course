### Data Wrangling

```python
# Drop the row where values are not NaN
indexNames = peking_weather[peking_weather['type'] == 'PM2.5_24h'].index
peking_weather.drop(indexNames, inplace = True)
indexNames = peking_weather[peking_weather['type'] == 'PM10_24h'].index
peking_weather.drop(indexNames, inplace = True)
peking_weather.head()
```

```python
for column_name in peking_weather.columns:
    peking_weather[column_name].replace(np.nan, 0, inplace=True)
```
